# 사회 연결망 분석(SNA·Social Network Analysis)

## 강의노트에 제공된 데이터들을 활용한 사회 연결망 분석

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

### 노드와 엣지 수동코딩

```python
G = nx.Graph()
```

```python
G.add_edge('정재정', '아이유', weight=0.3)
G.add_edge('아이유', '박명수', weight=0.6)
G.add_edge('정재정', '박명수', weight=0.9)
G.add_edge('박명수', '유재석', weight=1.2)
G.add_edge('아이유', '유재석', weight=1.5)
```

```python
nx.draw(G, with_labels=True, weighted=True) 
```

```python
G.nodes
```

### 노드 추가

```python
G.add_node('김구라') # 노드를 추가해본다.
```

```python
nx.draw(G, with_labels=True, weighted=True) # 외딴 곳에 연결되지 않은 김구라가 생성되었다.
```

### 엣지 추가

```python
G.edges()
```

```python
G.add_edge('김구라', '아이유', weight=2)
```

```python
nx.draw(G, with_labels=True, weighted=True) # 엣지 추가를 통해 김구라를 아이유와 연결했다.
```

### 네트워크 시각화

```python
nx.draw(G, with_labels=True, font_size=5, font_color='black', edge_color='black', node_color='red')
```

```python
nx.draw_random(G, with_labels=True) # 랜덤으로 그리기
```

#### 연결 중심성

```python
dict(G.degree())
```

## .csv 파일 불러와서 네트워크 분석하기

```python
featuring = pd.read_csv('C:/Users/jjj/Desktop/BA/featuring.csv') # 피처링 파일을 판다스의 read_csv를 통해 불러옵니다.
```

```python
GG = nx.from_pandas_edgelist(featuring,'from',"to",create_using = nx.DiGraph()) # from에서 to로 향하는 엣지리스트를 설정합니다.
```

```python
nx.draw_random(GG, with_labels=True) # 랜덤으로 네트워크 분석도를 그립니다.
```
